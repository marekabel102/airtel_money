```python
import matplotlib.pyplot as plt
```

```python

x = [0,2,4,6,8]
y = [0,4,16,36,64]

fig,ax = plt.subplots()
ax.plot(x,y,marker='o',label = 'Data points')

ax.set_title('Basic Components of Matplotlib figure')
ax.set_xlabel('x-Axis')
ax.set_ylabel('y-Axis')
plt.show()
```

```python
fruits = ('dates','oranges','mangoes','apples')
sales = (400,350,200,50)
plt.bar(fruits,sales)
plt.title('my fruits')
plt.xlabel('fruits')
plt.ylabel('sales')
plt.show()
```

```python
x=[3,1,3]
y = [3,2,1]

plt.plot(x,y)
plt.title('Line chart')

plt.legend(['Line'])
plt.show()
```

```python
# Pie Chart

cars = ('Audi','Jaguar','Ford','BMW','Toyota','VW','mazda')
data =(27,50,90,67,45,76,79,35)
fig = plt.figure(figsize = (10,7))
plt.pie(data,labels = cars)
plt.legend
plt.show()
```
